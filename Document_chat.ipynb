```python
import cassio
from datasets import load_dataset
from langchain.llms import OpenAI
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores.cassandra import Cassandra
from langchain.indexes.vectorstore import VectorStoreIndexWrapper

import warnings
warnings.filterwarnings('ignore')
```

```python
! pip install PyPDF2
```

```python
from PyPDF2 import PdfReader
pdfreader = PdfReader('Product Note_NipponIndiaLargeCapFund-Dec2023.pdf')

raw_text = ''
for i,pages in enumerate(pdfreader.pages):
    content = pages.extract_text()
    raw_text += content
```

```python
print(raw_text)
```

```python
astra_db_app_token = '********'
astra_db_id = '*************'
cassio.init(token=astra_db_app_token, database_id=astra_db_id)

openapi_key = '*****************'
llm = OpenAI(openai_api_key = openapi_key)

from langchain.embeddings import SentenceTransformerEmbeddings
embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_split = CharacterTextSplitter(
    separator = '\n',
    chunk_size = 800,
    chunk_overlap=200,
    length_function=len,
)
texts = text_split.split_text(raw_text)
```

```python
print(texts)
```

```python
astra_vector_store = Cassandra(
    embedding=embeddings,
    table_name='talk_to_pdf_1',
    session=None,
    keyspace=None,
)
astra_vector_store.add_texts(texts)
astra_vector_index = VectorStoreIndexWrapper(vectorstore=astra_vector_store)
```

```python
print(texts)
```

```python
astra_vector_index.query('what is the allocation % of HDFC bank ?',llm=llm)
```
